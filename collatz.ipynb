### old functions:

```python
# Outputs the number of steps it takes an input integer, n, to reach one
def steps_to_one(n):
    count = 0;
    j = n;
    while n != 1:
        if n%2 == 0:
            n = n/2;
            count += 1;
        elif n%2 != 0:
            n = (n * 3) + 1;
            count += 1;
    return [j, count]

# Outputs the even/odd pattern of a given number
def even_odd(n):
    count = 0;
    j = n;
    e_o_string = "";
    while n != 1:
        if n%2 == 0:
            n = n/2;
            count += 1;
            e_o_string += "E";
        elif n%2 != 0:
            n = (n * 3) + 1;
            count += 1;
            e_o_string += "O";
    return [j, e_o_string]

# Outputs input number, count of steps to one, and even/odd pattern of a given number
# this is the steps_to_one and even_odd functions put together
def collatz_first_iteration(n):
    count = 0;
    j = n;
    e_o_string = "";
    while n != 1:
        if n%2 == 0:
            n = n/2;
            count += 1;
            e_o_string += "E";
        elif n%2 != 0:
            n = (n * 3) + 1;
            count += 1;
            e_o_string += "O";
    return [j, count, e_o_string]
```

## current function

```python
def collatz(n):
    count = 0;
    j = n;
    e_o_string = "";
    even_count = 0;
    odd_count = 0;
    while n != 1:
        if n%2 == 0:
            n = n/2;
            count += 1;
            e_o_string += "E";
            even_count += 1;
        elif n%2 != 0:
            n = (n * 3) + 1;
            count += 1;
            e_o_string += "O";
            odd_count += 1;
    return [j, count, even_count, odd_count, e_o_string]
```
